```python
import pandas as pd
import numpy as np
```

```python
# Załadowanie danych
import pandas_datareader as web

df = web.DataReader(name='MAZN', data_source='stooq')

apple = web.DataReader(name='AAPL', data_source='stooq')
amazon = web.DataReader(name='AMZN', data_source='stooq')
google = web.DataReader(name='GOOGL', data_source='stooq')
uber = web.DataReader(name='UBER', data_source='stooq')
```

```python
apple.info()
apple.head()
```

```python
apple.columns =  ['apple_' + col.lower() for col in apple.columns]
amazon.columns =  ['amazon_' + col.lower() for col in amazon.columns]
google.columns =  ['google_' + col.lower() for col in google.columns]
uber.columns =  ['uber_' + col.lower() for col in uber.columns]
```

# Konkatenacja danych

```python
df = pd.concat(objs = [apple, amazon, google, uber], axis = 1)
```
