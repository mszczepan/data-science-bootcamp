# Załadowanie danych

```python
# Załadowanie danych
import pandas_datareader as web

df = web.DataReader(name='META', data_source='stooq')
```

# Zapis/Odczyt z pliku CSV

```python
df.to_csv('dane.csv')
df_nov = df[(df.index.month == 11) & (df.index.year == 2019) ]
df_nov
```

```python
df_nov.to_csv('META_nov.csv')
```
